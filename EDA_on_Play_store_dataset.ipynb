COUNT PLOT BY USING SNS(SEABORN)LIBRARY

```python
#we are plot the countplot with respect to the "Type" column
sns.countplot(x = df["Type"])
```

```python
sns.countplot(x = df["Content Rating"])
```

```python
# categorical columns graph with the help of matplotlib
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold',
alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):#range how many types run
    plt.subplot(2, 2, i+1)#size of the plot 2 x 2 and i+1 means it will increase the size
    sns.countplot(x=df[category[i]],palette="Set2")#plot the actual plot
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
